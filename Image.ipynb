# New Section

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd
import numpy as np

fig,ax = plt.subplots()

head= plt.Circle((50, 60),10,color='black',fill=False)
ax.add_artist(head) #head

#eyes
eye1 = patches.Ellipse((45, 61), 2,4, angle = 0,color='black', fill=True)
ax.add_artist(eye1)
eye2 = patches.Ellipse((55, 61), 2,4, color='black', fill=True)
ax.add_artist(eye2)

#smile
smile = patches.Arc((50, 57.5),6.5, 6.5,theta1=180, theta2=0,color='black')
ax.add_artist(smile)

#body
body = plt.Line2D([50,50], [50,20], color='black',linewidth=2.5)
ax.add_line(body)

#hands
lh = plt.Line2D([50,30], [40,30], color='black',linewidth=1.5)
rh = plt.Line2D([50,70], [40,30], color='black',linewidth=1.5)
ax.add_line(lh)
ax.add_line(rh)

#legs
ll = plt.Line2D([50,40], [20,0],color='black',linewidth=1.5)
rl = plt.Line2D([50,60], [20,0],color='black',linewidth=1.5)
ax.add_line(ll)
ax.add_line(rl)

#sun
sun = plt.Circle((80, 80), 6, color='yellow', fill=True)
ax.add_artist(sun)

#sunrays
for i in range(0,360,30):
    x= 9*np.cos(np.radians(i))
    y= 9*np.sin(np.radians(i))
    ray= plt.Line2D([80,80 + x],[80,80 + y],color='yellow')
    ax.add_line(ray)

for i in range(0,360,10):
    x =7.5*np.cos(np.radians(i))
    y =7.5*np.sin(np.radians(i))
    ray = plt.Line2D([80,80 + x], [80,80 + y], color='yellow')
    ax.add_line(ray)

#clouds
cloud1 =plt.Circle((15,85),8,color='skyblue')
ax.add_artist(cloud1)

cloud2= plt.Circle((25,90),8,color='skyblue')
ax.add_artist(cloud2)

cloud3 = plt.Circle((35,85), 8,color='skyblue')
ax.add_artist(cloud3)

#set axes
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)
ax.set_xticks([0, 20, 40, 60, 80, 100])
ax.set_yticks([0, 20, 40, 60, 80, 100])

plt.grid(True)
plt.show()
```
